# Nettoyage et traitement des données (2 points)

Votre DataFrame contient normalement des dates ou des chaines de caractères, il se peut aussi que vous ayez à retraiter des colonnes, effectuez 2 traitements de colonnes différents dans les 2 cellules ci-dessous

```python
# Ecrivez votre code ci-dessous

#On remarque qu'il y a des espaces unitiles dans les noms des variables donc on les renommes

df.rename(columns= {"Life expectancy ": "Life_expectancy", "Adult Mortality": "Adult_Mortality", 
                    "infant deaths":"infant_deaths", "percentage expenditure":"percentage_expenditure",
                    "Hepatitis B":"Hepatitis_B","Measles ":"Measles"," BMI ":"BMI",  "under-five deaths ":"under-five_deaths", 
                    "Total expenditure":"Total_expenditure","Diphtheria ":"Diphtheria", " HIV/AIDS":"HIV/AIDS",
                    " thinness  1-19 years":"thinness_10-19_years",
                    " thinness 5-9 years":"thinness_5-9_years",
                    "Income composition of resources":"Income_composition_of_resources"}, inplace=True)
          
```

```python
# Ecrivez votre code ci-dessous

# On a voulue procéder au cas par cas et supprimer les valeurs manquantes de Life_expectancy, 
# ensuite approchés les autres valeurs manquante par leurs moyennes ou mediane,
# cependant, selon la source des données, la plupart des données manquantes concernaient la population, l’hépatite B et le PIB.
# et proviennent de pays moins connus comme vanuatu, Tonga, Togo, Cabo Verde, etc.
# Il était difficile de trouver toutes les données pour ces pays et,
#  par conséquent, il a été décidé d’exclure ces pays de l’ensemble de données.

df= df.dropna()

df.isnull().sum()
```
